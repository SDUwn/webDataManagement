```python
from selenium import webdriver
import time
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains

def get_urls(url,xpath):
    driver=webdriver.Chrome()
    driver.get(url)
    ele=driver.find_elements_by_xpath(xpath)
    teacher_urls=[]
    for item in ele:
        teacher_urls.append(item.get_attribute("href"))
    return teacher_urls
```

```python
'''
get_urls_tag方法获取该学院教师的主页链接的列表
'''
def get_urls_tag(url,xpath):
    driver=webdriver.Chrome()#启动Chrome浏览器
    driver.get(url)#打开学院官网的教师名单页
    time.sleep(1)#等待一秒以确保信息加载完整（不等待则存在信息获取不完整的情况）
    ele=driver.find_element_by_xpath(xpath).find_elements_by_tag_name("a")#找到xpath后根据tag获取每位教师的主页链接
    teacher_urls=[]#存放每位教师的主页链接
    for item in ele:
        teacher_urls.append(item.get_attribute("href"))
    return teacher_urls
```

```python
'''
to_dataframe方法将数据转为dataframe，方便后续数据操作与整理
'''
import pandas
def to_dataframe(data):
    data=pandas.DataFrame(data)#将list类型的数据转为dataframe
    data=data[data['name']!=""]#去除name为空字符串的记录
    return data
```

```python
import pymysql
from sqlalchemy import create_engine
'''
save_to_mysql方法将数据导入到mysql数据库里
'''
def save_to_mysql(data,table_name):
    engine = create_engine("mysql+pymysql://{}:{}@{}/{}?charset={}".format('root', 'welcome2019', '127.0.0.1:3306', 'WebDataOfTeachers','utf8mb4'))
    con = engine.connect()#创建连接
    data.to_sql(name=table_name, con=con, if_exists='append', index=False)#将dataframe导入到mysql数据库的“teachers”表中。
```

```python
import json
import pymongo
def save_to_mongodb(data,collection_name):
    myclient = pymongo.MongoClient("mongodb://localhost:27017/")
    mydb = myclient["WebData"]
    mycol = mydb[collection_name]

    mylist =json.loads(data.T.to_json()).values()
    x = mycol.insert_many(mylist)

    # 输出插入的所有文档对应的 _id 值
    print(x.inserted_ids)
    # myclient.close()
```

### 上交计算机

```python
'''
在爬取每个教师主页时，大部分教师主页的信息并不是全部具备的，并且确实的属性不一，
有的缺失办公地址，有的缺失邮箱，等等，
因此针对每个属性获取时使用try except，确保在无法获取某一属性值时能获取到该教师的其他信息。
'''
#使用get_urls_tag方法获取教师主页链接列表
teacher_urls=get_urls_tag("http://www.cs.sjtu.edu.cn/Faculty.aspx","//*[@id=\"ctl00\"]/div[5]/div[2]/div/div[3]/div[2]")
teachers=[]#用于存放所有教师的信息
for teacher_url in teacher_urls:#对每一个教师主页链接操作
    try:
        teacher={}#用于存放每位教师的信息
        driver.get(teacher_url)#打开当前的教师主页
        try:
            #获取基本信息
            info=driver.find_element_by_xpath("//*[@id=\"Form1\"]/div[5]/div[2]/div/div[3]/div[1]").text
            #基本信息以“\n”分隔，获取一个基本信息属性的列表
            info=(info.split("\n"))
        except:
            info=""
        try:
            info_1=info[0].split(' ')
            info_1=[x.strip() for x in info_1 if x.strip() != '']#对数据进行规范，去除多余的空格等
            teacher['name']=info_1[0]+info_1[1]#获取名字
            teacher["position"]=info_1[2]#获取职称
        except:
            teacher['name']=""
            teacher['position']=""
        try:#获取邮箱
            teacher["email"]=info[4][5:]
        except:
            teacher["email"]=""
        try:#获取办公地址
            teacher["address"]=info[3][5:]+info[5][5:]
        except:
            teacher["address"]=""
        try:#获取研究领域
            #根据xpath获取信息，并将换行转换成空格以便数据规整
            teacher["workfield"]=driver.find_element_by_xpath("//*[@id=\"con_one_1\"]").text.replace("\n"," ")
        except:
            teacher["workfield"]=""
        
        try:   
            #获取教育经历
            #在当前页面不能获取到教育经历信息，使用selenium可以模仿人的动作进行爬取，
            #因此用selenium的actionchain模仿点击“教育经历”一栏，再爬取此时才会出现的教育经历信息数据。
            btn = driver.find_element_by_id('one2')
            action =ActionChains(driver)
            action.click(btn)
            action.perform()
            info_2=driver.find_element_by_id("con_one_2").text
            teacher["intro"]=info_2
            #获取工作经验
            #步骤与上述“教育经历”类似，不再赘述。
            btn = driver.find_element_by_id('one3')
            action = ActionChains(driver)
            action.click(btn)
            action.perform()
            info_2=driver.find_element_by_id("con_one_3").text
            #简历信息为教育经历与工作经验的结合
            teacher["intro"]=teacher["intro"]+info_2
        except:
            teacher["intro"]=""
        teachers.append(teacher)#将该教师信息添加到教师列表中
    except:
        teachers.append(teacher)
        continue
data=to_dataframe(teachers)#将获取的教师列表信息转成dataframe
save_to_mysql(data,"teachers")#存入mysql数据库
```
