```python
import requests
import gzip
```

```python
with open("list.csv","r",encoding='utf8')as f:
    l=f.read().splitlines()
```

```python
count=29
for i in l[28:]:
    id_=i.split('/')[-1]
    faa_name=id_+'_protein.faa.gz'
    gbff_name=id_+'_genomic.gbff.gz'
    r=requests.get(i+'/'+faa_name)
    with open('./data/'+faa_name,'wb') as f:
        f.write(r.content)
    r=requests.get(i+'/'+gbff_name)
    with open('./data/'+gbff_name,'wb') as f:
        f.write(r.content)
    print(count)
    count+=1
```

```python
import os
import time 
count=1
for i in filelist[0:]:
    start=time.time()
    os.system("blastp -task blastp-fast -query .\\data\\"+i+"_protein.faa -out .\\res\\"+i+".blast -db .\\base\\cogdb -outfmt 6 -evalue 1e-5  -num_threads 8")
    end=time.time()
    print('No'+str(count)+" "+i+"      timeused:"+str(end-start))
    count+=1
```
